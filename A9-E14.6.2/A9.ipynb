```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import binom

# function that computes posterior distribution
def posterior(x, n, q_values):
    likelihood = binom.pmf(x, n, q_values) #P(x | q)

    prior = np.ones_like(q_values) #P(q), which is uniform in [0, 1]

    posterior_unnormalized = likelihood * prior #P(q|x) = P(x|q) * P(q)

    return posterior_unnormalized / np.sum(posterior_unnormalized) #posterior normalized

#DATA
q_val = np.linspace(0, 1, 100)

x_1 = 80  # spam emails
x_2 = 15
n = 100  # emails
n_3 = 500

posterior_1 = posterior(x_1, n, q_val)
posterior_2 = posterior(x_2, n, q_val)
posterior_3 = posterior(x_1, n_3, q_val)

#PLOT
plt.figure(figsize=(14, 6))

plt.subplot(1, 3, 1)
plt.plot(q_val, posterior_1, color='blue')
plt.title('Posterior distribution with x = 80, n = 100')
plt.xlabel('$q$')
plt.ylabel('$P(q|x)$')
plt.ylim(0, 0.25)
plt.grid(True)

plt.subplot(1, 3, 2)
plt.plot(q_val, posterior_2, color='green')
plt.title('Posterior distribution with x = 15, n = 100')
plt.xlabel('$q$')
plt.ylabel('$P(q|x)$')
plt.ylim(0, 0.25)
plt.grid(True)

plt.subplot(1, 3, 3)
plt.plot(q_val, posterior_3, color='red')
plt.title('Posterior distribution with x = 80, n = 500')
plt.xlabel('$q$')
plt.ylabel('$P(q|x)$')
plt.ylim(0, 0.25)
plt.grid(True)

plt.tight_layout()
plt.show()
```
